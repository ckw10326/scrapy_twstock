##用request取得原始碼

```python
import requests

stock_num = 2408
url = r"https://www.istock.tw/stock/" + str(stock_num) + r"/chips-long" #股權分散表
#print(url)
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

##使用BS4處理html本文

```python
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(r.text, 'html.parser')
```

###定址在想要的表格

```python
html_soup.find('table', class_='ecostyle1 table table-striped m-b-none')
```

###抓表頭

```python
my_table_head = html_soup.find("thead")
html_soup.find("thead")
```

```python
titles = html_soup.find("thead")('th')
html_soup.find("thead")('th')
```

```python
headers = []
for element in titles:
    headers.append(element.text)
print(headers)
```

先自製表頭

```python
enumerate(headers)
for i, j in enumerate(headers):
    print(i, j)
```

```python
mytable_dict = {"日期":[], "零股(%, 人數)":[], "10張以下(%, 人數)":[], "100-200張(%, 人數)":[], "200-400張(%, 人數)":[], "400-1000張(%, 人數)":[], "1000張以上(%, 人數)":[], "總股東人數":[], "成較量":[], "收盤":[], "開盤":[], "最高":[], "最低":[]}
```

###抓內容

```python
table = html_soup.find('table', class_='ecostyle1 table table-striped m-b-none')
for eachrow in table('tr'):
  print(eachrow)
```

```python
table = html_soup.find('table', class_='ecostyle1 table table-striped m-b-none')
for eachrow in table('tr')[2:]:#不取前面兩行表頭
  my_data = eachrow(['tr','td'])

  #當那行的長度不對時跳過
  if len(my_data) == 13:
    for indexnum, datas in enumerate(mytable_dict.keys()):
      #print(indexnum, datas)
      #print(my_data[indexnum].text.strip())
      mytable_dict[datas].append( my_data[indexnum].text.strip() )

  #mytable_dict[item].append( my_data[idx].text.strip() )
```
